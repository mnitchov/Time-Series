```python
import numpy as np
import pandas as pd
from pathlib import Path
%matplotlib inline

import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

```python
# Currency pair exchange rates for CAD/JPY
cad_jpy_df = pd.read_csv(
    Path("cad_jpy.csv"), index_col="Date", infer_datetime_format=True, parse_dates=True
)
cad_jpy_df.head()
```

```python
# Trim the dataset to begin on January 1st, 1990
cad_jpy_df = cad_jpy_df.loc["1990-01-01":, :]
cad_jpy_df.head()
```

```python
# Plot just the "Price" column from the dataframe:
cad_jpy_df.Price.plot()
```

```python
# Question: Do you see any patterns, long-term and/or short?  
# Answer: The data shows that between 1992-1996 there was a downtrend. 1996-2008 shows a higher low was made then reaches a pinnacle at 2008. It then takes naother dip and levels off around 2020. We can
# see that there is an overall downtrend amongst the data.
```

```python
import statsmodels.api as sm

# Apply the Hodrick-Prescott Filter by decomposing the exchange rate price into two separate series:
ts_noise, ts_trend = sm.tsa.filters.hpfilter(cad_jpy_df['Price'])
ts_trend.rename(columns={'Price':'Trend'},inplace=True)
ts_trend.head()
```

```python
ts_noise.rename(columns={'Price':'Noise'},inplace=True)
ts_noise.head()
```

```python
# Create a dataframe of just the exchange rate price, and add columns for "noise" and "trend" series from above:
cad_jpy_df = cad_jpy_df[['Price']].copy()
cad_jpy_df['Trend'] = ts_trend
cad_jpy_df['Nosie'] = ts_noise
cad_jpy_df.head()
```

```python
# Plot the Exchange Rate Price vs. the Trend for 2015 to the present
cad_jpy_df.plot(y=['Price', 'Trend'],figsize=[20,10],title='Price vs Trend Plot',legend=True)
```

```python
# Question: Do you see any patterns, long-term and/or short? 
# Answer: We can see short term fluctuations where the blue line dips below the yellow or is higher than it. Long term trend still shows the exchange rate value is decreasing
```

```python
 # Plot the Settle Noise
ts_noise.plot(figsize=[20,10],title='Noise',legend=True)
```

```python
# Using futures Settle Returns, estimate an ARMA model

# ARMA: Create an ARMA model and fit it to the returns data. Note: Set the AR and MA ("p" and "q") parameters to p=2 and q=1: order=(2, 1).
# Output the ARMA summary table and take note of the p-values of the lags. Based on the p-values, is the model a good fit (p < 0.05)?
# Plot the 5-day forecast of the forecasted returns (the results forecast from ARMA model)
```

```python
# Create a series using "Price" percentage returns, drop any nan"s, and check the results:
# (Make sure to multiply the pct_change() results by 100)
# In this case, you may have to replace inf, -inf values with np.nan"s

returns = (cad_jpy_df[["Price"]].pct_change() * 100)
returns = returns.replace(-np.inf, np.nan).dropna()
returns.tail()
```

```python
import statsmodels.api as sm
from statsmodels.tsa.arima_model import ARMA

# Estimate and ARMA model using statsmodels (use order=(2, 1))
model = ARMA(returns.values, order=(2,1))

# Fit the model and assign it to a variable called results
results = model.fit()
```

```python
# Output model summary results:
results.summary()
```

```python
# Plot the 5 Day Returns
pd.DataFrame(results.forecast(steps=5)[0]).plot(title="5 Day Forecast of Returns")
```

```python
# Question: Based on the p-value, is the model a good fit?
# Answer: No since there are values over 0.05
```

```python
from statsmodels.tsa.arima_model import ARIMA

# Estimate and ARIMA Model:
# Hint: ARIMA(df, order=(p, d, q))

model = ARIMA(cad_jpy_df['Price'], order=(5, 1, 1))

# Fit the model
results = model.fit()
```

```python
# Output model summary results:
results.summary()
```

```python
# Plot the 5 Day Forecast

pd.DataFrame(results.forecast(steps=5)[0]).plot(title="5 Day Forecast of Returns")
```

```python
# Question: What does the model forecast will happen to the Japanese Yen in the near term? 
# Answer: We can see that there is an overall downtrend between the USD and yen in the short term over a 5 day period
```

```python
# Rather than predicting returns, let's forecast near-term volatility of Japanese Yen futures returns. Being able to accurately predict volatility will be extremely useful if we want to trade in derivatives or quantify our maximum loss.

# Using futures Settle Returns, estimate an GARCH model

# GARCH: Create an GARCH model and fit it to the returns data. Note: Set the parameters to p=2 and q=1: order=(2, 1).
# Output the GARCH summary table and take note of the p-values of the lags. Based on the p-values, is the model a good fit (p < 0.05)?
# Plot the 5-day forecast of the volatility.

from arch import arch_model

# Estimate a GARCH model:
model = arch_model(returns['Price'], mean="Zero", vol="GARCH", p=2,q=1)
# Fit the model
results = model.fit()
```

```python
# Summarize the model results

results.summary()
```

```python
# Find the last day of the dataset

last_day = returns.index.max().strftime('%Y-%m-%d')
last_day
```

```python
# Create a 5 day forecast of volatility

forecasts_horizon = 5

# Start the forecast using the last_day calculated above

forecasts = results.forecast(start=last_day, horizon=forecasts_horizon)
forecasts
```

```python
# Annualize the forecast

intermediate = np.sqrt(forecasts.variance.dropna() * 252)
intermediate.head()
```

```python
 # Transpose the forecast so that it is easier to plot
final = intermediate.dropna().T
final.head()
```

```python
# Plot the final forecast
final.plot(title='5 Day Forecast of Volatility', legend=True)
```

```python
# Question: What does the model forecast will happen to volatility in the near term?
# Answer: Expected to increase
```
